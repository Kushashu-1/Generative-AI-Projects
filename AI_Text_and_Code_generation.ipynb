```python
from transformers import pipeline
```

**Generator Model**

```python
generator = pipeline('text-generation' , model = 'EleutherAI/gpt-neo-2.7B')
```

**Generate Text Using Promt**

```python
def apply(promt):
  res = generator(promt , max_length = 50 , do_sample = True , temperature = 0.9)
  return res
```

**Prompt and Output**

```python
promt = "What is Machine"
res = apply(promt)[0]['generated_text']
```

```python
with open('sample.txt','w') as f:
  f.write(res)
```
